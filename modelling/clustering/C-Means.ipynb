```python
import pandas as pd
import numpy as np
import skfuzzy as fuzz

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=100_000, centers=7, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X and Y')
plt.show()
```

```python
%%time

n_clusters = 7

cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(X.T, n_clusters, 2, error=0.001, maxiter=20000, init=None)

cluster_labels = np.argmax(u, axis=0)

#print(cluster_labels)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X and Y with Cluster Labels')
plt.show()
```

```python
import numpy as np

# Calculate the membership values for each data point
membership_values = np.max(u, axis=0)

# Print the membership values
print(membership_values)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=membership_values)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X and Y with Membership Values')
plt.colorbar(label='Membership Values')
plt.show()
```
